```python
import pandas as pd
```

```python
filename = 'EPL Results 1993-2018.csv'
```

```python
df = pd.read_csv(filename)
```

```python
print(df)
```

```python
df.fillna(value='0')
```

```python
club_names_list = df.drop_duplicates(subset=['HomeTeam'])
```

```python
print(club_names_list)
```

```python
club_names_list.sort_values(by='HomeTeam', ascending=True)
```

```python
club_names_list = club_names_list.filter(['HomeTeam'], axis=1)
```

```python
print(club_names_list)
```

```python
club_names_list = club_names_list.reset_index(drop=True)
```

```python
print(club_names_list)
```

```python
Home_results = df.groupby(["HomeTeam", "FTR"]).size().unstack(fill_value=0)
```

```python
print(Home_results)
```

```python
Home_results.sort_values(by='H', ascending=False)
```

```python
games_played = df.groupby('HomeTeam')['Season'].count()
```

```python
print(games_played)
```

```python
games_played = games_played.to_frame()
```

```python
print(games_played)
```

```python
games_played = games_played['Season'].sort_values(ascending=False)
```

```python
print(games_played)
```

```python
Merge_Table = pd.merge(Home_results, games_played, on='HomeTeam')
```

```python
print(Merge_Table)
```

```python
Merge_Table = Merge_Table.reset_index()
```

```python
print(Merge_Table)
```

```python
output = pd.DataFrame(columns=['HomeTeam', 'Result'])
```

```python
for index, row in Merge_Table.iterrows():
    result = row['H'] / row['Season'] * 100
    output = output.append({'HomeTeam': row['HomeTeam'], 'Result': result}, ignore_index=True)
```

```python
print(output)
```

```python
import numpy as np
```

```python
row_mean = np.mean(output, axis=0)
```

```python
print(row_mean)
```
